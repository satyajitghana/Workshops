```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_img(img, cmap='gray'):
    plt.figure(figsize=(6, 6))
    plt.imshow(img, cmap=cmap)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
myimagegray = cv2.imread('knuth.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
myimage = cv2.imread('knuth.jpg')
```

```python
drawed_im = cv2.rectangle(myimage, (10,10), (50,50), (0, 255, 0), thickness=2)
```

```python
show_img(drawed_im)
```

```python
myimagegray = cv2.imread('knuth.jpg', cv2.IMREAD_GRAYSCALE)
myimage = cv2.imread('knuth.jpg')
```

# Hardcascades

```python
face_learner = cv2.CascadeClassifier(r"C:\Users\shadowleaf\Anaconda3\envs\tflow\Library\etc\haarcascades\haarcascade_frontalface_default.xml")
```

```python
face_coords = face_learner.detectMultiScale(myimagegray, 1.2, 5) # compare 30% data at a atime
face_coords
```

```python
[[x, y, w, h]] = face_coords
```

```python
drawed_im = cv2.rectangle(myimagegray, (x,y), (x+w,y+h), (0, 255, 0), thickness=2)
```

```python
show_img(drawed_im)
```

```python
iroids_color = cv2.imread('IROIDS_MASK.jpg')
iroids = cv2.imread('IROIDS_MASK.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
faces = face_learner.detectMultiScale(iroids, 1.2, 5)
faces
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(iroids_color,  (x, y), (x+w, y+h), (0, 0, 255), thickness=5)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(iroids_color, cv2.COLOR_BGR2RGB))
plt.xticks([])
plt.yticks([])
plt.show()
```
