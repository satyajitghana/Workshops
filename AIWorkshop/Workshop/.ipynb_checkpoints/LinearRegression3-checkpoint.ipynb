```python
import pandas as pd
```

```python
mydata = pd.read_csv('Advertising.csv')
```

```python
mydata.head()
```

```python
X_input, Y_output = mydata.iloc[:, 1:4].values.reshape(200, 3), mydata.iloc[:, 4].values.reshape(200, 1)
```

```python
X_train, Y_train = X_input[:140], Y_output[:140]
X_test, Y_test   = X_input[140:], Y_output[140:]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Yp = model.predict(X_test)
Ya = Y_test
```

```python
model.score(X_test, Y_test)
```

```python
model.coef_, model.intercept_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.plot(X_test[:, 0], Y_test, 'o', markersize = 10)
plt.plot(X_test[:, 0], Yp, 'rv', markersize = 10)
plt.xlabel('TV')
plt.legend(['Ya', 'Yp'])
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(X_test[:, 1], Y_test, 'ro', markersize = 10)
plt.plot(X_test[:, 1], Yp, 'bv', markersize = 10)
plt.xlabel('Radio')
plt.legend(['Ya', 'Yp'])
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(X_test[:, 2], Y_test, 'mo', markersize = 10)
plt.plot(X_test[:, 2], Yp, 'yv', markersize = 10)
plt.xlabel('Newspaper')
plt.legend(['Ya', 'Yp'])
plt.grid()
```
